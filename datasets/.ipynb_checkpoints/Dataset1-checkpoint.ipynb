```python

benchmarks = {
    'Hospitals_ratio': 1.0,  
    'Elementary_schools_ratio': 1.0,  
    'Secondary_schools_ratio': 1.0,   
    'Gym_ratio': 1.0,           
    'Park_ratio': 1.0,          
    'Rest_coffee_ratio': 3.0,   
    'Mosques_ratio': 1.0              
}


unit_sizes = {
    'Hospitals_ratio': 10000,            
    'Elementary_schools_ratio': 800,     
    'Secondary_schools_ratio': 800,      
    'Gym_ratio': 5000,                   
    'Park_ratio': 2386,                  
    'Rest_coffee_ratio': 1000,           
    'Mosques_ratio': 1000                       
}


population_columns = {
    'Elementary_schools_ratio': 'Children',
    'Secondary_schools_ratio': 'Youth',
    'Hospitals_ratio': 'Total population of the neighborhood',
    'Gym_ratio': 'Total population of the neighborhood',
    'Park_ratio': 'Total population of the neighborhood',
    'Rest_coffee_ratio': 'Total population of the neighborhood',
    'Mosques_ratio': 'Males above 10 years'
}


def calculate_needed_units(row, factor, benchmark, unit_size, population_column):
    actual_ratio = row[factor]
    population = row[population_column]
    
    if actual_ratio < benchmark:
        
        additional_units = ((benchmark - actual_ratio) * population) / unit_size
        return max(int(additional_units), 0)  
    return 0  


for factor, benchmark in benchmarks.items():
    unit_size = unit_sizes[factor]
    population_column = population_columns[factor]
    column_name = factor.replace('_ratio', '_needed')  
    
    
    df[column_name] = df.apply(lambda row: calculate_needed_units(row, factor, benchmark, unit_size, population_column), axis=1)


```

```python
weights = {
    'Elementary_schools_ratio': 0.2,
    'Secondary_schools_ratio': 0.2,
    'Hospitals_ratio': 0.2,
    'Gym_ratio': 0.1,
    'Park_ratio': 0.1,
    'Rest_coffee_ratio': 0.1,
    'Mosques_ratio': 0.1
}


def calculate_total_score_as_percentage(row, weights):
    total_score = 0
    for factor, weight in weights.items():
        
        if factor == 'Rest_coffee_ratio':
            normalized_ratio = row[factor] / 5 
            total_score += normalized_ratio * weight
        else:
            total_score += row[factor] * weight  
    return total_score * 100  


df['Total_Score_Percentage'] = df.apply(lambda row: calculate_total_score_as_percentage(row, weights), axis=1)

```

```python
start_year = 2024
end_year = 2030
years = list(range(start_year, end_year + 1))


build_rates = {
    'Mosques_needed': 4,  
    'Secondary_schools_needed': 1,  
    'Elementary_schools_needed': 1,  
    'Hospitals_needed': 1,  
    'Gym_needed': 2,  
    'Park_needed': 3,  
    'Rest_coffee_needed': 20  
}


build_periods = {
    'Mosques_needed': 1,  
    'Secondary_schools_needed': 2,  
    'Elementary_schools_needed': 1,  
    'Hospitals_needed': 3,  
    'Gym_needed': 1,  
    'Park_needed': 1,  
    'Rest_coffee_needed': 1  
}


def calculate_yearly_plan(instances_needed, build_rate_per_year, years, build_period=None):
    yearly_plan = {year: 0 for year in years} 
    remaining_instances = instances_needed     

    for year in years:
        
        if build_period:
            if (year - start_year) % build_period != 0:
                continue  

        if remaining_instances > 0:
            if remaining_instances >= build_rate_per_year:
                yearly_plan[year] = build_rate_per_year
                remaining_instances -= build_rate_per_year
            else:
                yearly_plan[year] = remaining_instances
                remaining_instances = 0  

    return yearly_plan


def flatten_plan(df, factor_column, build_rate, build_period):
    flat_data = []
    for _, row in df.iterrows():
        neighborhood = row['Neighborhood']
        instances_needed = row[factor_column]
        plan = calculate_yearly_plan(instances_needed, build_rate, years, build_period)
        for year, count in plan.items():
            
            if count > 0:
                flat_data.append({'Neighborhood': neighborhood, 'Year': year, 'Factor': factor_column, 'Instances_Built': count})
    return pd.DataFrame(flat_data)


mosques_flat = flatten_plan(df, 'Mosques_needed', build_rates['Mosques_needed'], build_periods['Mosques_needed'])
secondary_schools_flat = flatten_plan(df, 'Secondary_schools_needed', build_rates['Secondary_schools_needed'], build_periods['Secondary_schools_needed'])
elementary_schools_flat = flatten_plan(df, 'Elementary_schools_needed', build_rates['Elementary_schools_needed'], build_periods['Elementary_schools_needed'])
hospitals_flat = flatten_plan(df, 'Hospitals_needed', build_rates['Hospitals_needed'], build_periods['Hospitals_needed'])
gyms_flat = flatten_plan(df, 'Gym_needed', build_rates['Gym_needed'], build_periods['Gym_needed'])
parks_flat = flatten_plan(df, 'Park_needed', build_rates['Park_needed'], build_periods['Park_needed'])
rest_coffee_flat = flatten_plan(df, 'Rest_coffee_needed', build_rates['Rest_coffee_needed'], build_periods['Rest_coffee_needed'])


final_df = pd.concat([mosques_flat, secondary_schools_flat, elementary_schools_flat, hospitals_flat, gyms_flat, parks_flat, rest_coffee_flat])
```
